# D. PREPROCESSING & MODELING

## 1. melakukan preprocessing data berdasarkan variable dataset  yang ingin di pakai untuk melakukan modeling

```python
# mengisi nilai yang hilang dengan string kosong
def preprocessing_data(df):
  df['need'] = df['need'].fillna('')
  df['training_name'] = df['training_name'].fillna('')

#menggabungkan nilai dari atribut - atribut yang akan digunakan
  df['combined_attributes'] = df['need']+ '' + df['training_name']

  return df
```

## 2. melakukan inisialisasi objek menggunakan Tfidfvectorizer dan membuat matriks TF - IDF dari atribut gabungan

```python
# Inisialisasi object Tfidfvecktorizer
def calculate_tfidf_matrix(data):
  tfidf_vectorizer = TfidfVectorizer(stop_words = 'english')

# Membuat matriks TF - IDF dari atribut gabungan
  tfidf_matrix = tfidf_vectorizer.fit_transform(data['combined_attributes'])

  return tfidf_matrix
```

## 3. Membuat function untuk mendapatkan rekomendasi berdasarkan hasil inputan yang dimasukkan

```python
def get_recommendations(need, cosine_sim, data, search_recommendations=len(df), parameter_cosine_similarity=0.1):
    idx = data.loc[(data['need'] == need)].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:search_recommendations+1]
    company_indices = [i[0] for i in sim_scores]

    return data[['training_name']].iloc[company_indices].drop_duplicates()
```

## 4. menginisialisasikan kembali dataset dan melakukan perhitugann score similarity dan memasukan inputan untuk melakukan rekomendasi

```python
# Tidak memanggil dataset lagi karena sudah dipanggil sebelumnya
# df = pd.read_csv('AI_SALES FITUR 6.csv')  # Gantilah 'your_dataset.csv' dengan nama dataset sesuai kebutuhan
df = preprocessing_data(df)

# Tetap menggunakan dataset yang sudah dipanggil sebelumnya
tfidf_matrix = calculate_tfidf_matrix(df)

# Menghitung skor kesamaan kosinus dengan parameter_cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Mendapatkan rekomendasi untuk training_needs tertentu
need = 'Hukum Bisnis Indonesia Consult'
parameter_cosine_similarity = 0.1  # Atur nilai parameter_cosine_similarity sesuai kebutuhan
recommendations = get_recommendations(need, cosine_sim, df, parameter_cosine_similarity=parameter_cosine_similarity)[:10]
```
